## Ejercicios Teóricos

1. ¿Qué diferencia hay entre una estrategia de búsqueda Informada y una estrategia de búsqueda No Informada?

Las estrategias de búsqueda no informada no tienen información adicional sobre los estados más allá de lo que nos dice el problema, y sólo distingue un estado objetivo del que no lo es. En cambio, las estrategias de búsqueda informada sí ofrecen información completa sobre los estados sucesores y el estado objetivo, lo que permite encontrar la solución óptima respecto a las estrategias de búsqueda no informada.

2. ¿Qué es una heurística y para qué sirve?

Una heurística es una fucnción que nos ayuda a estimar el camino más barato para llegar de un estado cualquiera a un estado objetivo. Las funciones heurísticas son la forma más común de transimir el conocimiento adicional de problema al algoritmo de búsqueda. Si n es un nodo objetivo, entonces h(n) = 0.

3. Analice cuántos estados posibles puede adoptar el Mundo de la Aspiradora si existieran 3 habitaciones.

El agente (aspiradora) puede estar en 3 localizaciones posibles, cada una de las cuales puede contener suciedad o no. Así, hay 3*2^3=24 estados posibles.

4. ¿Es posible que un algoritmo de búsqueda no tenga solución?

Sí, es posible que el algoritmo de búsqueda no tenga solución. Puede pasar que el estado objetivo no exista. Si la estrategia de búsqueda es informada, se puede verificar que exista antes de implementar el algoritmo de búsqueda. También puede pasar que el árbol de búsqueda tenga una profundidad infinita y, si se implementa un algoritmo de búsqueda primero en profundidad y la solución está en otra rama diferente, entonces no va a encontrar la solución.
